```
import pandas as pd

ARQUIVO_OCEANS = 'dados_oceans14_final.csv'
ARQUIVO_STATUS = 'dados_status_invest-2020_06_15.csv'
df_oceans = pd.read_csv('Oceans14/'+ARQUIVO_OCEANS, encoding='utf-8')
df_status = pd.read_csv('Status Invest/'+ARQUIVO_STATUS, encoding='utf-8')
```

```
df_status['ano'] = 2020
```

Listando as colunas exclusivas de cada base de dados

```
colunas_exclusivas_oceans = list(set(df_oceans.columns) - set(df_status.columns))
```

```
colunas_exclusivas_status = list(set(df_status.columns) - set(df_oceans.columns))
```

Separando da Status os dados das empresas que tem na Oceans14

```
ticks_oceans = list(df_oceans['tick'].unique())
```

```
df_status = df_status[df_status['tick'].isin(ticks_oceans)].copy()
```

```
df_oceans = df_oceans.set_index('tick')
df_status = df_status.set_index('tick')
```

## Separando apenas os dados de 2020 da Oceans14

```
df_oceans_2020 = df_oceans[df_oceans['ano'] == 2020]
```

```
df_oceans_2020[colunas_exclusivas_oceans]
```

Concatenando dados Status com da Oceans14

```
#pd.concat([df_oceans_2020[colunas_exclusivas_oceans], df_status], axis=1)
df_2020_final = df_status.join(df_oceans_2020[colunas_exclusivas_oceans])
```

```
df_2020_final.fillna(0, inplace=True)
df_2020_final
```

```
df_2020_final.to_csv('dados_2020_final.csv', encoding='utf-8')
```

## Juntando as duas bases, mesmo que em anos anteriores, varias colunas estejam faltando

Filtra os dados da Oceans14 em ano = 2020, e junta com dados da Status, que só tem 2020

```
df_oceans_status_2020 = df_oceans[df_oceans['ano'] == 2020][colunas_exclusivas_oceans].join(df_status).copy()
```

Junta os dados completos de 2020 com os outros da Oceans14.
Isso gera varios NaN, esperado.

```
df_oceans_status = pd.concat([df_oceans[df_oceans['ano'] != 2020], df_oceans_status_2020])
```

```
df_oceans_status.to_csv('dados_oceans_status.csv', encoding='utf-8')
```
